```python

def add_rows(votes, votings):
    vote_counts = votes.groupby(['VoteId', 'Vote']).size().unstack(fill_value=0)
    
    vote_counts = vote_counts.rename(columns={
        1: "Yes_votes",
        2: "No_votes",
        3: "Abstentions_votes",
        4: "Absent_votes",
        5: "Didnt_vote_votes",
        6: "Motivated_votes",
        0: "Not_a_mep_votes"
    })
    
    votings = votings.merge(vote_counts, on='VoteId', how='left').fillna(0)
    
    votings = votings.astype({col: int for col in vote_counts.columns if col in votings.columns})

    return votings
```

```python
votings6 = add_rows(vote6,votings6)
votings7 = add_rows(vote7,votings7)
votings8 = add_rows(vote8,votings8)
votings9 = add_rows(vote9,votings9)
```

```python
def makeproblem(votings):
    problem = votings.copy()
    return problem[
        (problem["Yes"].astype(int) != problem["Yes_votes"].astype(int)) |
        (problem["No"].astype(int) != problem["No_votes"].astype(int)) |
        (problem["Abs"].astype(int) != problem["Abstentions_votes"].astype(int))
    ]
```

```python
problem6 = makeproblem(votings6)
problem7 = makeproblem(votings7)
problem8 = makeproblem(votings8)
problem9 = makeproblem(votings9)
```
